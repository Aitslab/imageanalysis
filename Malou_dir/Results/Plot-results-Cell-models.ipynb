```python
import skimage
import skimage.io
import os
import matplotlib.pyplot as plt
from IPython.display import Markdown, display
from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

```python
channel = "CELL/"

resultfolder = "/home/malou/Master/results/Resultfolder/" + channel + "results/"
models = sorted(os.listdir(resultfolder))
```

```python
for model in models:
    fig_path = resultfolder + model + "/results/figs/"
    if os.path.isdir(fig_path):
        model_name = model.split("_")
        model_name = "{} {}".format(model_name[0],model_name[1]) 
        display(Markdown("# <div align='center'>{}</div>".format(model_name)))
        im = skimage.io.imread(fig_path + "csv_plot.png")
        plt.imshow(im)
        plt.tick_params(axis='both', labelsize=0, length = 0)
        plt.show()
        figs = os.listdir(fig_path)
        for fig in figs:
            if fig != "csv_plot.png":
                im = skimage.io.imread(fig_path + fig)
                plt.figure(figsize = (20,10))
                plt.imshow(im)
                plt.tick_params(axis='both', labelsize=0, length = 0)
                plt.show()
```
