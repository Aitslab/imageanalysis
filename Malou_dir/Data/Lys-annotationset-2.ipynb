```python
masks_path = '/media/malou/Seagate Expansion Drive/Malou_Master/Master/data/0_lysosome_ch_images/lys-annotationset2/annotations/'
norm_path = '/media/malou/Seagate Expansion Drive/Malou_Master/Master/data/0_lysosome_ch_images/lys-annotationset2/d1_images/'
masks = os.listdir(masks_path)
norm_im = os.listdir(norm_path)


for image in masks:
    
    mask = skimage.io.imread(masks_path + image)
    norm = skimage.io.imread(norm_path + image)
    display(Markdown("# <div align='center'>{}</div>".format(image)))
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20,10))
    #fig.suptitle(masks[0])
    ax1.imshow(mask)
    ax2.imshow(norm)
    ax1.tick_params(axis='both', labelsize=0, length = 0)
    ax2.tick_params(axis='both', labelsize=0, length = 0)
    plt.show()
```
